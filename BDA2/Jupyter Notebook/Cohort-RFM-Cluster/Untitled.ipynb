```python
import pandas as pd

# Create two dataframes representing the two tables
table1 = pd.DataFrame({'id': [1, 2, 3],
                       'name': ['John', 'Alice', 'Bob']})

table2 = pd.DataFrame({'id': [2, 3, 4],
                       'age': [25, 30, 35]})

# Join the two tables based on a common column (id)
merged_table = pd.merge(table1, table2, on='id')

# Print the merged table
print(merged_table)
```
