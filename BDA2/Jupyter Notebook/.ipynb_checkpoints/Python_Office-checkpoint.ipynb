```python
wb = load_workbook(path+'report_2023.xlsx')
sheet = wb['Report']
file_name='report_2023.xlsx'
# cell references (original spreadsheet)
min_column = wb.active.min_column
max_column = wb.active.max_column
min_row = wb.active.min_row
max_row = wb.active.max_row
month_and_extension = file_name.split('_')[1]
```

```python
min_column,max_column,min_row,max_row,month_and_extension
```

```python
sheet['B8'] = '=SUM(B6:B7)'
sheet['B8'].style = 'Currency'
```

```python
wb.save(path+'report_2023.xlsx')
```

```python
# adding a chart
barchart = BarChart()
data = Reference(sheet, min_col=min_column+1, max_col=max_column, min_row=min_row, max_row=max_row) #including headers
categories = Reference(sheet, min_col=min_column, max_col=min_column, min_row=min_row+1, max_row=max_row) #not including headers
barchart.add_data(data, titles_from_data=True)
barchart.set_categories(categories)
sheet.add_chart(barchart, "B12") #location chart
barchart.title = 'Sales by Product line'
barchart.style = 2 #choose the chart style

wb.save(path+'report_2023.xlsx')
```

```python
# applying formulas
# first create alphabet list as references for cells
alphabet = list(string.ascii_uppercase)

excel_alphabet = alphabet[0:max_column] #note: Python lists start on 0 -> A=0, B=1, C=2. #note2 the [a:b] takes b-a elements

excel_alphabet
```

```python
# sum in columns B-G
for i in excel_alphabet:
    if i!='A':
        sheet[f'{i}{max_row+1}'] = f'=SUM({i}{min_row+1}:{i}{max_row})'
        sheet[f'{i}{max_row+1}'].style = 'Currency'
wb.save(path+'report_2023.xlsx')   
```

```python
sheet[f'{excel_alphabet[0]}{max_row+1}'] = 'Total'
wb.save(path+'report_2023.xlsx')
```

```python
def as_text(value):
    if value is None:
        return ""
    return str(value)

for column_cells in sheet.columns:
    length = max(len(as_text(cell.value)) for cell in column_cells)
    sheet.column_dimensions[column_cells[0].column_letter].width = length
wb.save(path+'report_2023.xlsx')
```

```python
# getting month name
month_name = month_and_extension.split('.')[0]

# formatting the report
sheet['A1'] = 'Sales Report'
sheet['A2'] = month_name.title()
sheet['A1'].font = Font('Arial', bold=True, size=20)
sheet['A2'].font = Font('Arial', bold=True, size=10)

wb.save(path+'report_2023.xlsx')
```

## send emails though Outlook

```python
import win32com.client

def send_outlook(msg, report, to_address):
    outlook = win32com.client.Dispatch("Outlook.Application")
    mail = outlook.CreateItem(0)
    mail.To = to_address
    mail.Subject = f'Your {report} is generated'
    mail.Body = msg
#     mail.Attachments.Add(attachment)
    mail.Send()
```

```python
import pandas as pd
import openpyxl
from openpyxl import load_workbook
from openpyxl.styles import Font
from openpyxl.chart import BarChart, Reference
import string
path=r'C:\Business_Data_Analysis\Jupyter Notebook\python automation/'
def automate_excel(file_name):
   """The file name should have the following structure: sales_month.xlsx"""
   # read excel file
   excel_file = pd.read_csv(file_name)
   # make pivot table
   report_table = excel_file.pivot_table(index='Gender', columns='Product line', values='Total', aggfunc='sum').round(0)
   # splitting the month and extension from the file name
   month_and_extension = file_name.split('_')[1]
   print(month_and_extension)
   # send the report table to excel file
#    report_table.to_excel(f'report_{month_and_extension}', sheet_name='Report', startrow=4)
   report_table.to_excel(path+'report_2023.xlsx',sheet_name='Report',startrow=4)
   # loading workbook and selecting sheet
   wb = load_workbook(path+'report_2023.xlsx')
#    wb = load_workbook(f'report_{month_and_extension}')
   sheet = wb['Report']
   # cell references (original spreadsheet)
   min_column = wb.active.min_column
   max_column = wb.active.max_column
   min_row = wb.active.min_row
   max_row = wb.active.max_row
   # adding a chart
   barchart = BarChart()
   data = Reference(sheet, min_col=min_column+1, max_col=max_column, min_row=min_row, max_row=max_row) #including headers
   categories = Reference(sheet, min_col=min_column, max_col=min_column, min_row=min_row+1, max_row=max_row) #not including headers
   barchart.add_data(data, titles_from_data=True)
   barchart.set_categories(categories)
   sheet.add_chart(barchart, "B12") #location chart
   barchart.title = 'Sales by Product line'
   barchart.style = 2 #choose the chart style
   # applying formulas
   # first create alphabet list as references for cells
   alphabet = list(string.ascii_uppercase)
   excel_alphabet = alphabet[0:max_column] #note: Python lists start on 0 -> A=0, B=1, C=2. #note2 the [a:b] takes b-a elements
   # sum in columns B-G
   for i in excel_alphabet:
       if i!='A':
           sheet[f'{i}{max_row+1}'] = f'=SUM({i}{min_row+1}:{i}{max_row})'
           sheet[f'{i}{max_row+1}'].style = 'Currency'
   sheet[f'{excel_alphabet[0]}{max_row+1}'] = 'Total'
   # getting month name
   month_name = month_and_extension.split('.')[0]
   # formatting the report
   sheet['A1'] = 'Sales Report'
   sheet['A2'] = month_name.title()
   sheet['A1'].font = Font('Arial', bold=True, size=20)
   sheet['A2'].font = Font('Arial', bold=True, size=10)
   wb.save(path+'report_py.xlsx')
   return
file_name=path+"supermarket_sales.csv"
automate_excel(file_name)
```

```python
import openpyxl
from docx import Document
from docx.shared import Inches
from openpyxl.chart import BarChart, Reference, LineChart
from io import BytesIO

# Enter the file names for your Excel and Word documents
excel_file_name = "example.xlsx"
word_file_name = "example.docx"

# Enter the name of the Excel worksheet you want to use
worksheet_name = "Report"

# Open the Excel file and select the worksheet
wb = openpyxl.load_workbook(path+'report_2023.xlsx')
ws = wb[worksheet_name]



# Get the number of rows and columns in the worksheet
num_rows = ws.max_row
num_cols = ws.max_column

# Find the first non-empty row in the worksheet
first_row = 1
while all(cell.value is None for cell in ws[first_row]):
    first_row += 1

# Create a new Word document
document = Document()
document.add_heading('Document Title', 0)

p = document.add_paragraph('A plain paragraph having some ')
p.add_run('bold').bold = True
p.add_run(' and some ')
p.add_run('italic.').italic = True

document.add_heading('Heading, level 1', level=1)
document.add_paragraph('Intense quote', style='Intense Quote')

document.add_paragraph(
    'first item in unordered list', style='List Bullet'
)
document.add_paragraph(
    'first item in ordered list', style='List Number'
)


# Add a table to the document and set its style
table = document.add_table(rows=1, cols=num_cols)
table.style = 'Table Grid'

# Add the column headers to the table
hdr_cells = table.rows[0].cells
for i in range(num_cols):
    hdr_cells[i].text = str(ws.cell(row=first_row, column=i+1).value)

# Add the rows from the Excel worksheet to the table
for row in ws.iter_rows(min_row=first_row+1, values_only=True):
    row_cells = table.add_row().cells
    for i in range(num_cols):
        row_cells[i].text = str(row[i])

        

# Enter the file name for your PNG image
image_file_name = path+"test.png"

# Add a paragraph to the document
document.add_paragraph("Here is an example PNG image:")

# Add the PNG image to the document
document.add_picture(image_file_name, width=Inches(4))        
        
        
        
# Save the Word document
document.save(word_file_name)
```

# extract text from PPT

```python
from pptx import Presentation

prs = Presentation(r"C:\Users\ABC\Desktop\1112.pptx")

# text_runs will be populated with a list of strings,
# one for each text run in presentation
text_runs = []

for slide in prs.slides:
    for shape in slide.shapes:
        if not shape.has_text_frame:
            continue
        for paragraph in shape.text_frame.paragraphs:
            for run in paragraph.runs:
                text_runs.append(run.text)
```

### Create the HTML Report
#### Parameters 

```python
import pandas as pd
import smtplib
import matplotlib.pyplot as plt

from email.mime.image import MIMEImage
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.message import EmailMessage
from email.mime.base import MIMEBase

import codecs
```

```python
dict_days = dict(zip(['MON', 'TUE', 'WED', 'THU', 'FRI', 'SAT', 'SUN'], 
                     ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
                    ))
WEEK = 'WEEK-1'
```

#### Import Data

```python
# Volumes per day
df_day = pd.read_csv('volumes per day.csv')
df_day
```

```python
df_plot = df_day[df_day['WEEK'] ==WEEK].copy()

''' Calculate KPIs for Comments'''
# Lines per orders
df_plot['LINES/ORDER'] = df_plot['LINES']/df_plot['ORDERS']
avg_ratio = '{:.2f} lines/order'.format(df_plot['LINES/ORDER'].mean())
max_ratio = '{:.2f} lines/order'.format(df_plot['LINES/ORDER'].max())
print(avg_ratio)
print(max_ratio)

# Maximum Day Lines
busy_day = dict_days[df_plot.set_index('DAY')['LINES'].idxmax()]
max_lines = '{:,} lines'.format(df_plot['LINES'].max())
print(busy_day)
print(max_lines)

# Total Workload
total_lines = '{:,} lines'.format(df_plot['LINES'].sum())
print(total_lines)

# Bar Plot: Orders/Lines
fig, ax = plt.subplots(figsize=(14, 7))
df_plot.plot.bar(figsize=(8, 6), edgecolor='black', x='DAY', y=['ORDERS', 'LINES'], 
                                           color=['tab:blue', 'tab:red'], legend= True, ax = ax)
plt.xlabel('DAY', fontsize = 12)
plt.title('Workload per day (Lines/day)', fontsize = 12)
plt.show()
```

#### Save the visual

```python
# save plot
filename = 'visual' + ".png"
path_plot = filename
fig.savefig(path_plot, dpi=fig.dpi)
```

```python
# Files Name
file_img = 'visual.png'
file_header =  "header_img.png"
# SMTP Server (example of gmail)
smtp_server = 'smtp.google.com'
smtp_port = 587
# Settings
from_mail = 'f.tableau.4@gmail.com'
from password import *
to_mail = "f.tableau.4@gmail.com"

# Message Object
msg = MIMEMultipart()
msg['Subject'] = 'Workload Report for {}'.format(WEEK)
msg['From'] = from_mail
msg['To'] = ', '.join([from_mail, to_mail])
msg.preamble = 'Workload Report for {}'.format(WEEK)

# Open the visual
with open(file_img, 'rb') as fp:
    img = MIMEImage(fp.read())
    img.add_header('Content-Disposition', 'attachment', filename=file_img)
    img.add_header('X-Attachment-Id', '0')
    img.add_header('Content-ID', '<0>')
    fp.close()
    msg.attach(img)

# Open the header
with open(file_header, 'rb') as fp:
    img = MIMEImage(fp.read())
    img.add_header('Content-Disposition', 'attachment', filename=file_header)
    img.add_header('X-Attachment-Id', '1')
    img.add_header('Content-ID', '<1>')
    fp.close()
    msg.attach(img)

# Attach the HTML email
f = codecs.open("report.html", 'r')
string = f.read()

# Add the images in the HTML report
html_string = string.replace("IMG_HTML", "cid:0")
html_string = html_string.replace("HEADER_HTML", "cid:1")

# Add the insights
html_string = html_string.replace("WEEK", WEEK)
html_string = html_string.replace("total_lines", total_lines)
html_string = html_string.replace("busy_day", busy_day)
html_string = html_string.replace("max_lines", max_lines)
html_string = html_string.replace("avg_ratio", avg_ratio)
html_string = html_string.replace("max_ratio", max_ratio)

# Attach the HTML page
msg.attach(MIMEText(html_string, 'html', 'utf-8'))

from os.path import basename
from email.mime.application import MIMEApplication
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.utils import COMMASPACE, formatdate

files=['test.pptx']
for f in files or []:
        with open(f, "rb") as fil:
            part = MIMEApplication(
                fil.read(),
                Name=basename(f)
            )
        # After the file is closed
        part['Content-Disposition'] = 'attachment; filename="%s"' % basename(f)
        msg.attach(part)
```

# excel to PDF

```python
import os
import openpyxl
from reportlab.lib.pagesizes import letter
from reportlab.lib.units import inch
from reportlab.pdfgen import canvas

# Enter the file name for your Excel file
excel_file_name = "merge1.xlsx"

# Enter the file name for your PDF file
pdf_file_name = "example.pdf"

# Open the Excel file using openpyxl
workbook = openpyxl.load_workbook(excel_file_name, read_only=True)
worksheet = workbook.active

# Set the size of the PDF canvas and the bottom margin
canvas_width = 20 * inch
canvas_height = 11 * inch
bottom_margin = 0.5 * inch

# Set the position of the Excel table within the PDF file
x, y = 0.5 * inch, canvas_height - 0.5 * inch

# Set the current page and the maximum number of rows per page
current_page = 1
max_rows_per_page = 25

# Set the width and spacing of the columns in the Excel table
column_width = 2 * inch
column_spacing = 0.5 * inch

# Create the PDF file and add the first page
pdf = canvas.Canvas(pdf_file_name, pagesize=(canvas_width, canvas_height))
pdf.setFont("Helvetica", 12)
pdf.drawCentredString(canvas_width/2, canvas_height - 0.75 * inch, "Excel Table")

# Loop through the rows and columns of the Excel file
for row in worksheet.iter_rows():
    # Check if the current row exceeds the maximum number of rows per page
    if (row[0].row - 1) % max_rows_per_page == 0 and row[0].row != 1:
        # If it does, add a new page to the PDF file and update the position of the Excel table
        pdf.showPage()
        current_page += 1
        y = canvas_height - 0.5 * inch
        
    for cell in row:
        # Check if the cell is not empty
        if cell.value is not None:
            # Get the cell value and add it to the PDF file
            cell_value = str(cell.value)
            pdf.drawString(x + (cell.column - 1) * (column_width + column_spacing), y - (cell.row - (current_page - 1) * max_rows_per_page) * 0.25 * inch, cell_value)

    # Update the position of the Excel table
    y -= 0.25 * inch

# Save the PDF file
pdf.save()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

excel_file = pd.read_csv(r"C:\Users\ABC\Desktop\python automation\supermarket_sales.csv")
excel_file[['Gender', 'Product line', 'Total']]
report_table = excel_file.pivot_table(index='Gender',columns='Product line',values='Total',aggfunc='sum').round(1).reset_index()
```

```python
fig, ax = plt.subplots(2,1)
# fig, ax = plt.subplots(3,1, figsize=(10,4))
# hide axes
fig.patch.set_visible(False)
ax[0].axis('off')
ax[0].axis('tight')

df = report_table
ax[0].set_title("Sales Report 2021", fontsize=18,y=1,pad=-12)
table=ax[0].table(cellText=df.values,colLabels=df.columns, loc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.5, 1.5)
table.auto_set_column_width(col=list(range(len(df.columns))))
plt.show()
fig.savefig('test.pdf', bbox_inches='tight', dpi=600)
```

```python
df.plot.bar(figsize=(15,9),ax=ax[1])

fig.tight_layout()
plt.show()
fig.savefig('test.pdf', bbox_inches='tight', dpi=600)
# fig.savefig('test.html', bbox_inches='tight', dpi=600)
```

# EXCEL TO  HTML

```python
import plotly

from plotly.subplots import make_subplots
import plotly.graph_objects as go
df = report_table
figure = make_subplots(
        rows = 2, cols  =1,
        specs=[[{"type":"table"}],[{"type" : "bar"}]]
    )

figure.add_trace(go.Table(
    header=dict(values=list(df.columns),
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[df[col] for col in df.columns],
               fill_color='lavender',
               align='left')), row=1, col=1
)
figure.add_trace(go.Bar(
    x=df.index,
    y=df['Electronic accessories']
),row=2,col=1)

figure.update_layout(height=800, width=800, title_text="Sales Report 2021")
plotly.offline.plot(figure, filename='figure.html',validate=False)

```
